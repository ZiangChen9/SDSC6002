# 1.Import

```python
import torch
import gpytorch
import botorch
import matplotlib.pyplot as plt
from botorch.test_functions.synthetic import ThreeHumpCamel,Hartmann
from numpy.ma.core import negative
from torch.quasirandom import SobolEngine
from botorch.sampling.normal import SobolQMCNormalSampler

from botorch.models import SingleTaskGP
from botorch.acquisition import qKnowledgeGradient, qLogNoisyExpectedImprovement
from botorch.acquisition.predictive_entropy_search import qPredictiveEntropySearch
from gpytorch.mlls import ExactMarginalLogLikelihood
from botorch.fit import fit_gpytorch_mll
from tqdm.notebook import tqdm
from botorch.models.transforms import Normalize, Standardize
from botorch.optim import optimize_acqf

from gpytorch.means import LinearMean
from gpytorch.kernels import ScaleKernel, RBFKernel
import warnings
import csv
```

# 2.Test Function: Valley-Shaped: ThreeHumpCamel

```python
tkwargs = {
    "dtype": torch.float64,
    "device": torch.device("cuda:0"),
}
d=2
bounds=torch.tensor([[-5.0, -5.0],[5.0,  5.0]],**tkwargs)
optimizer =torch.tensor([0.0, 0.0], **tkwargs)
optimal_value=0.0
test_f= ThreeHumpCamel(negate = True)
noise =0.5
def f4(x,noise):
  y=test_f(x)
  y=test_f(x)+noise*torch.randn_like(y)
  return y.unsqueeze(-1)
def f(x,noise):
  return f4(x,noise)
```

# 3.BayesOpt Loop: ThreeHumpCamel+qKG+ConstantMean+RBFKernel

```python
best_values=[]
result_reals=[]
runs=1
for i in tqdm(range(runs)):
  # Set SEED
  SEED=i
  torch.manual_seed(SEED)
  torch.cuda.manual_seed_all(SEED)
  d=2

  # Initial xt and yt
  Init_num=4*d
  sobol= SobolEngine(dimension=d, scramble=True)
  xt = sobol.draw(Init_num).to(dtype=torch.float64,device=torch.device("cuda:0"))
  xt[:,0]=xt[:,0]*(bounds[1][0]-bounds[0][0])+bounds[0][0]
  xt[:,1]=xt[:,1]*(bounds[1][1]-bounds[0][1])+bounds[0][1]
  yt=f(xt,noise)
  yr=f(xt,0)

  # Records
  result_real=[val.item() for val in yr]
  result_noise=[val.item() for val in yt]
  best_value=[-max(result_real)]

  # Budget
  budget=15*d

  # Sampler
  sampler = SobolQMCNormalSampler(torch.Size([1024]))

  # Mean & Kernel
  mean_fn = gpytorch.means.ConstantMean()
  kernel_fn = gpytorch.kernels.ScaleKernel(gpytorch.kernels.RBFKernel())

  xt_1=xt
  # Loop
  for j in tqdm(range(budget)):
    # GP Model
    gp_1=SingleTaskGP(train_X=xt,
                      train_Y=yt,
                      input_transform=Normalize(d=d),
                      outcome_transform=Standardize(m=1),
                      mean_module=mean_fn,
                      covar_module=kernel_fn)
    # Fit
    fit_gpytorch_mll(ExactMarginalLogLikelihood(gp_1.likelihood, gp_1))

    # Acquisition Function
    num_fantasies=128
    acf=qKnowledgeGradient(model=gp_1,num_fantasies=num_fantasies)

    # Second surrogate
    sobol_engine = SobolEngine(dimension=d, scramble=True, seed=SEED)
    sobol_samples = sobol_engine.draw(num_fantasies).to(**tkwargs)
    xt_2=xt_1
    yt_2=[]
    for k in range(Init_num):
      yt_2.append(acf(torch.cat((xt_2[k].view(1,d),sobol_samples),0)).item())
    yt_2=torch.tensor(yt_2,**tkwargs).view(Init_num,1)

    # inner
    sampler = SobolQMCNormalSampler(torch.Size([1024]))
    for inner in range(30):
      gp_2=SingleTaskGP(train_X=xt_2,
                        train_Y=yt_2,
                        input_transform=Normalize(d=2),
                        outcome_transform=Standardize(m=1))
      fit_gpytorch_mll(ExactMarginalLogLikelihood(gp_2.likelihood, gp_2))
      acf_2=qLogNoisyExpectedImprovement(model=gp_2,X_baseline=xt_2, sampler=sampler)
      candidate, acq_value = optimize_acqf(acf_2, bounds=bounds, q=1, num_restarts=10, raw_samples=50,)
      xt_2= torch.cat([xt_2, candidate], dim=0)
      yt_2_next= torch.tensor([[acf(torch.cat((candidate,sobol_samples),0)).item()]],**tkwargs).view(-1,1)
      yt_2=torch.cat([yt_2,yt_2_next],dim=0)
    yt_2_max,yt_2_index=torch.max(yt_2,dim=0)

    # next train_X
    candidate = xt_2[yt_2_index]
    candidate = candidate.to(**tkwargs)

    # List of train_X
    xt= torch.cat([xt, candidate], dim=0)

    # next train_Y & next real_Y
    yr_next = f(candidate,0)
    yt_next = f(candidate,noise)

    # List of train_Y
    yt = torch.cat([yt, yt_next])

    # List of train_Y & List of real_Y
    result_noise.append(yt_next.squeeze(-1).item())
    result_real.append(yr_next.squeeze(-1).item())
    best_value.append(-max(result_real))
  # Update Records
  result_reals.append(result_real)
  best_values.append(best_value)

  # Plot
  iter_num=[k for k in range(0, budget+1)]
  min=[optimal_value for k in range(0, budget+1)]
  if i%3 == 0:
    plt.figure(figsize=(8, 6))
    plt.xlabel("Number of evaluations")
    plt.ylabel("Best value found")
    plt.title(f"Runs:{i+1}-{i+1+2}")
    plt.plot(iter_num,min,'--',label='Optimal Value')
  plt.plot(iter_num,best_value,label=f'Run:{i+1}')
  plt.legend(loc='upper right',)

  # Print
#   print(result_real)
# print(result_reals)
with open('result_reals_ThreeHumpCamel+qKG+ConstantMean+RBF.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerows(best_values)
```
